```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets 
from sklearn import model_selection
from sklearn import preprocessing
from sklearn import svm
from sklearn import model_selection
from sklearn.datasets import samples_generator
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import *
from sklearn.ensemble import AdaBoostClassifier
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(np.insert(iris.data, 4, iris.target, axis=1), columns=iris.feature_names + ["flower"])
df.tail()
```

```python
df.plot()
```

```python
mlp_list = []
for alpha in 10.0 ** - np.arange(1, 7):
    print(alpha)
    mlp = MLPClassifier(alpha=alpha)
    mlp_list.append(mlp) 
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(preprocessing.scale(iris.data), iris.target)
```

```python
for mlp in mlp_list:
    mlp.fit(x_train, y_train)
```

```python
for mlp in mlp_list:
    print(mlp.score(x_test, y_test))
```

```python
for mlp in mlp_list:
    print(confusion_matrix(y_test, mlp.predict(x_test)))
```

```python
for mlp in mlp_list:
    print(recall_score(y_test, mlp.predict(x_test), average='micro'))
```

```python
for mlp in mlp_list:
    print(f1_score(y_test, mlp.predict(x_test), average='micro'))
```
