```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
```

```python
df = pd.read_csv('Var_2_groceries - groceries.csv')
df.head()
```

```python
items = []
for i in range(32):
    i+=1
    items+=(df["Item " + str(i)].unique().tolist())
    
items = set(items)
items = list(items)
items = [x for x in items if str(x) != 'nan']
```

```python
vals = []
for index, row in df.iterrows():
    labels = {}
    unmatch = list(set(items) - set(row))
    match = list(set(items).intersection(row))
    for u in unmatch:
        labels[u] = 0
    for m in match:
        labels[m] = 1
    vals.append(labels)
ap_df = pd.DataFrame(vals)
vals[0]
```

```python
freq_items = apriori(ap_df, min_support=0.01, use_colnames=True)
freq_items.head()
```

```python
rules = association_rules(freq_items, metric="lift", min_threshold=1)
rules [(rules['confidence']>=0.5)]
```

```python
ant = [list(val)[0] for val in rules['antecedents'].values]
con = [list(val)[0] for val in rules['consequents'].values]
res = pd.DataFrame(list(zip(ant,con)),columns=['antecedents', 'consequents'])
res
```
